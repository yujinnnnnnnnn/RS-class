```python
import numpy as np
import os.path
import matplotlib.pyplot as plt
import rasterio
from rasterio.plot import show
```

```python
fname_img='/content/무.jpg'
```

```python
os.path.isfile(fname_img)
```

```python
img = rasterio.open(fname_img)
```

```python
show(img)
```

```python
type(img)
```

```python
I = img.read()
```

```python
I.shape
```

```python
type(I)
```

```python
I = np.swapaxes(I, 0, 2)
```

```python
I.shape
```

```python
X = I.reshape((788*684, 3))
```

```python
X.shape
```

```python
I[0,1, :]
```

```python
X[1, :]
```

```python
I[1, 0, :]
```

```python
X[788, :]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca.fit(X)
```

```python
print(pca.components_)
```

```python
x1 = pca.components_[:,0]
x2 = pca.components_[:,1]
np.dot(x1, x1)
```

```python
np.dot(x1, x2)
```

```python
print(pca.explained_variance_)
```

```python
plt.plot(np.arange(3)+1, pca.explained_variance_, 'bo-')
```

```python
X_pca = pca.transform(X)
```

```python
X_pca.shape
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], marker='.', alpha=0.2)
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 2], marker='.', alpha=0.2)
```

```python
I_pca = X_pca.reshape(788, 684, 3)
```

```python
I_pca = np.swapaxes(I_pca, 0, 1)
```
